```python
from claspy.segmentation import BinaryClaSPSegmentation
from claspy.data_loader import load_tssb_dataset
import pandas as pd
import numpy as np
```

```python
# Read the text file into a DataFrame without header row and with specific column names
df = pd.read_csv("D:/Aitogy Projects/Drilling Identifier/data/lamhong2imu.txt", sep=' ', header=None, names=['timestamp', 'Gx', 'Gy', 'Gz', 'Ax', 'Ay', 'Az'])

# Display the DataFrame
# print(df.head)
df["Gx"]
```

```python
AccelX = np.array(df["Ax"].values, dtype=np.float64)
true_cps_AccelX = np.array([6274, 10790, 22675, 41758])

AccelY = np.array(df["Ay"].values, dtype=np.float64)
true_cps_AccelY = np.array([6274, 10790, 22675, 41758])

AccelZ = np.array(df["Az"].values, dtype=np.float64)
true_cps_AccelZ = np.array([6274, 10790, 22675, 41758])

GyroX = np.array(df["Gx"].values, dtype=np.float64)
true_cps_GyroX = np.array([6274, 10790, 22675, 41758])

GyroY = np.array(df["Gy"].values, dtype=np.float64)
true_cps_GyroY = np.array([6274, 10790, 22675, 41758])

GyroZ = np.array(df["Gz"].values, dtype=np.float64)
true_cps_GyroZ = np.array([6274, 10790, 22675, 41758])

clasp = BinaryClaSPSegmentation(n_segments=5, window_size="fft", excl_radius=50, validation=None)
```

```python
X = clasp.fit_predict(AccelX)
clasp.plot(gt_cps=true_cps_AccelX, heading="Segmentation of accelerator IMU data", ts_name="Ax", file_path="segmentationAx.png")
```

```python
print(X)
```

```python
Y = clasp.fit_predict(AccelY)
clasp.plot(gt_cps=true_cps_AccelY, heading="Segmentation of accelerator IMU data", ts_name="Ay", file_path="segmentationAy.png")
```

```python
print(Y)
```

```python
Z = clasp.fit_predict(AccelZ)
clasp.plot(gt_cps=true_cps_AccelZ, heading="Segmentation of accelerator IMU data", ts_name="Az", file_path="segmentationAz.png")
```

```python
print(Z)
```

```python
GX = clasp.fit_predict(GyroX)
clasp.plot(gt_cps=true_cps_GyroX, heading="Segmentation of gyroscope IMU data", ts_name="Gx", file_path="segmentationGx.png")
```

```python
GY = clasp.fit_predict(GyroY)
clasp.plot(gt_cps=true_cps_GyroY, heading="Segmentation of gyroscope IMU data", ts_name="Gy", file_path="segmentationGx.png")
```
